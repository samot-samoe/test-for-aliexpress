```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-ru")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-ru")
```

```python
def translate(text):
  
  tokenized_text = tokenizer.prepare_seq2seq_batch([text], return_tensors='pt')#,paddind = True)
  translation = model.generate(**tokenized_text)#(src_text, return_tensors="pt", padding=True))
  translated_text = tokenizer.batch_decode(translation, skip_special_tokens=False)[0]
  return translated_text

for i in range(100,199):
  print(textwrap.fill(translate(df['messages'][i])),end='\n------------------\n')
```
